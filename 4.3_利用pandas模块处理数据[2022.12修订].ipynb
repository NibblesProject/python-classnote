## 4.3.1 Series数据结构

Series（一维结构）：1个索引列+1个数据列。索引值默认是从0起递增的整数。  

列表，字典等都可以用来创建Series数据结构，与列表不同的是，Series的索引可以指定，类型可以是字符串类型。

```python
#创建Series对象
import pandas as pd
yw1=pd.Series([89,91,78,66,86,99])
print(yw1)
```

```python
import pandas as pd
yw2=pd.Series([89,91,78,66,86,99],index=["小红","小明","小李","李华","张三","李四"])
print(yw2)
```

```python
import pandas as pd
yw3=pd.Series({"小红":89,"小明":91,"小李":78,"李华":66,"张三":86,"李四":99})
print(yw3)
```

```python
#查看Series对象的值
for i in yw1:
    print(i)
```

```python
for i in yw2.index:
    print(i)
```

```python
for i in yw2.values:
    print(i)
```

```python
# 创建一个Series（变量名s3）：["小学"，"初中"，"高中"，"大学"]，并指定他们的index分别为7，12，15，18
import pandas as pd
s3=pd.Series(["小学","初中","高中","大学"],index=[7,12,15,18])
print(s3)

import pandas as pd
s3=pd.Series({"小学":7,"初中":12,"高中":15,"大学":18})
print(s3)
```

DataFrame是一种类似于关系表的表格型数据结构，DataFrame对象是一个二维表格, 其中每列中的元素类型必须一致，而不同的列可以拥有不同的元素类型。


*
